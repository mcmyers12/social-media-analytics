```python
# %load /Users/piorkja1/Documents/PycharmProjects/Platform_Access/RedditTopicModeler.py

__author__ = 'piorkja1'

# First Part is to Read Reddit and store comments
# Simple PRAW Script to Read SubRedit on Hopkins
#import praw 
import json
import gensim
import urllib2
from pprint import pprint as pp

# setup for tokenization and stopwords
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
from gensim import corpora, models, similarities
#from collections import defaultdict

stop_words = set(stopwords.words('english'))
stop_words.update(['.',  ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}', '#', '@']) # remove if you need punctuation

'''# collect JHU Posts from REDDIT
with open("redditprocess.txt", "w") as rpf:
    r = praw.Reddit(client_id = '1RmB7ChqHgjuZw', client_secret='xl04Tf2edeM6k_0hmnQrWFdmvrs', user_agent='me')
    for post in r.subreddit('JHU').comments(limit=500):
        rpf.write(post.body.encode('utf-8') + '\n')
        words = [i.lower() for i in wordpunct_tokenize(post.body.encode('utf-8')) if i.lower() not in stop_words]
        #print (words)
        all_tokens = ' '.join(words)
    #print (words)
    #dictionary = corpora.Dictionary(words)
    #print (dictionary)'''

# collect Instagram captions for posts hashtagged 'acroyoga'
with open('instagram-captions.txt', 'w') as igf:
    owners = []
    htag = 'acroyoga'
    search_url = 'https://www.instagram.com/explore/tags/' + htag + '/?__a=1'
    # search_url = 'https://api.instagram.com/tags/snowy/media/recent'

    contents = urllib2.urlopen(search_url).read()
    results = json.loads(contents)
    edges = results['graphql']['hashtag']['edge_hashtag_to_media']['edges']
    for edge in edges:
        captions = edge['node']['edge_media_to_caption']['edges']
        for caption in captions:
            text = caption['node']['text']
            igf.write(text.encode('utf-8') + '\n')
            words = [i.lower() for i in wordpunct_tokenize(text.encode('utf-8')) if i.lower() not in stop_words]
            all_tokens = ' '.join(words)

texts = words
# remove words that appear only once
#all_tokens = sum(texts, [])
tokens_once = set(words for words in set(all_tokens) if all_tokens.count(words) == 1)
#print(tokens_once)
texts = [[words for words in texts if words not in tokens_once]
         for words in all_tokens]
#print (texts)

    #file.write(text + '\n')

#    file.write(post.body + '\n')

# Setup for Document Matrix

#Setup gensim dictionary

dictionary = corpora.Dictionary(texts)


corpus = [dictionary.doc2bow(text) for text in texts]
corpora.MmCorpus.serialize('reddit.mm', corpus)
#print(corpus)

lda = gensim.models.LdaModel(corpus, id2word=dictionary, alpha='auto', num_topics=10)
for i in lda.show_topics():
    print (i)
#convert ot BOW vectors

import pyLDAvis.gensim

topic_vis = pyLDAvis.gensim.prepare(lda, corpus, dictionary)

pyLDAvis.display(topic_vis)





```
